```python
import requests
from datetime import datetime
from IPython.display import display
from bis2 import iucn
from bis2 import mlab
```

```python
bisDB = mlab.getDB("bis")
iucncache = bisDB["iucncache"]
```

```python
for iucnDoc in iucncache.find({"iucnSpecies":{"$exists":False}}):
    iucnSpeciesResult = requests.get(iucn.addToken(iucnDoc["searchURL"])).json()
    display (iucnSpeciesResult)
    print (iucncache.update_one({"_id":iucnDoc["_id"]},{"$set":{"speciesCacheDate":datetime.utcnow().isoformat(),"iucnSpecies":iucnSpeciesResult}},upsert=False))
    
```
